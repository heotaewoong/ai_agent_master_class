```python
# Movie Expert Agent 
import openai

url = "https://nomad-movies.nomadcoders.workers.dev"

client = openai.OpenAI() 

PROMPT = """
You are a Movie Expert Agent. You have access to the following functions:

1. get_popular_movies() - /movies에서 인기 영화를 가져옵니다.
2. get_movie_details(id) - /movies/:id에서 영화 정보를 가져옵니다.
3. get_movie_credits(id) - /movies/:id/credits에서 출연진 및 제작진을 가져옵

The base URL for all APIs is: https://nomad-movies.nomadcoders.workers.dev

Please answer only with the function name and the required arguments.
Example: get_movie_details(550)

"""

# 2. 테스트할 질문들입니다.
queries = [
    "지금 인기 있는 영화가 무엇인지 알려줘",
    "movie ID 550에 해당하는 영화가 무엇인지 알려줘",
    "movie ID 550에 해당하는 영화에 누가 출연하는지 알려줘"
] 

for query in queries: 
    response = client.chat.completions.create(
        model = "gpt-4o-mini", 
        messages = [
            {
                "role": "system",
                "content": PROMPT
            }
        ],
            {
                "role": 
            }    
    )

answer = response.choices[0].message.content
print(f"질문: {query}")
print(f"에이전트의 선택: {answer}")
print("-" * 30)
```

1. role: "system" vs role: "user" 차이점
결론부터 말씀드리면, user로 다 몰아서 넣어도 작동은 합니다! 하지만 system 역할을 쓰는 데는 중요한 이유가 있어요.

system (기획자/감독): AI에게 **"너는 이제부터 이런 규칙을 지키는 영화 전문가야"**라고 세계관과 규칙을 주입하는 역할이에요. 대화가 길어져도 이 규칙을 우선적으로 기억하려고 노력합니다.

user (손님/질문자): AI에게 **"지금 영화 하나 추천해줘"**라고 실제로 시키는 일(명령)을 의미합니다.

비유하자면?

system: 배우에게 "당신은 오늘부터 엄격한 탐정 역할을 맡았습니다"라고 대본을 주는 것.

user: 탐정에게 "이 범인이 누구죠?"라고 질문하는 것.

만약 모두 user에 넣으면, AI는 가끔 규칙(대본)과 질문을 헷갈려 하거나, 대화가 길어질 때 "아까 내가 정해준 규칙이 뭐였지?" 하고 까먹을 확률이 높아져요. 그래서 규칙은 system, **질문은 user**로 나누는 것이 표준입니다.
